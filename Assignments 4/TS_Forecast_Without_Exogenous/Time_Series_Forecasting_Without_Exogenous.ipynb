```python
import numpy as np
from pycaret.time_series import *
from pycaret.datasets import get_data
```

```python
import pandas as pd
data = pd.read_csv('hotel_prices_forecasting.csv')
data['date'] = pd.to_datetime(data['date'])
data.head()
```

```python
forecast_year = 12
forecast_fold = 3
```

```python
setup = setup(data, fh=forecast_year, fold=forecast_fold, target = 'hotel_price')
```

```python
setup.plot_model()
```

```python
setup.plot_model(plot = 'acf')
```

```python
setup.plot_model(plot = 'pacf')
```

```python
setup.plot_model(plot = 'periodogram')
```

```python
setup.plot_model(plot = 'fft')
```

```python
setup.plot_model(plot = 'diagnostics')
```

```python
best_model = setup.compare_models()
```

```python
created_model = setup.create_model('snaive')
```

```python
predicted_model = setup.predict_model(created_model)
```

```python
plot_model(created_model)
```

```python
tuned_model = setup.tune_model(created_model)
plot_model(tuned_model)
```

```python
finalized_model = setup.finalize_model(tuned_model)
setup.save_model(finalized_model, "finalized_model_time_series_without_exogenous_variables")
```

```python
loaded_model = setup.load_model("finalized_model_time_series_without_exogenous_variables")
loaded_model
```
