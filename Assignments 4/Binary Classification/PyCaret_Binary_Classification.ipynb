```python
import pandas as pd
data = pd.read_csv("emails.csv")
data.head()
```

```python
print(data.dropna())
```

```python
data.dropna(inplace=True)
from pycaret.classification import *
initial_setup = setup(data, target='Prediction', use_gpu = True)
```

```python
compare_models = compare_models()
```

```python
help(plot_model)
```

```python
logistic_regression = create_model('lr')
```

```python
plot_model(compare_models, plot = 'auc')
```

```python
plot_model(compare_models, plot = 'confusion_matrix')
```

```python
plot_model(compare_models, plot = 'feature')
```

```python
pred = predict_model(compare_models)
```
