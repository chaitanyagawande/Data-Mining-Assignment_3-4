```python
import pandas as pd
data = pd.read_csv('CreditCardTransaction.csv')
data.head()
```

```python
from pycaret.anomaly import *
```

```python
setup = setup(data, use_gpu = True)
```

```python
models()
```

```python
isolation_forest = create_model('iforest')
isolation_forest
```

```python
svm_model = create_model('svm')
```

```python
svm_model
```

```python
plot_model(isolation_forest, plot = 'tsne')
```

```python
plot_model(svm_model, plot = 'tsne')
```

```python
evaluate_model(isolation_forest)
```

```python
evaluate_model(svm_model)
```

```python
isolation_forest_anomalies = assign_model(isolation_forest)
isolation_forest_anomalies
```

```python
svm_model_anomalies = assign_model(svm_model)
svm_model_anomalies
```

```python
isolation_forest_pred = predict_model(isolation_forest, data=data)
isolation_forest_pred
```

```python
save_model(isolation_forest, 'isolation_forest_anomaly_detection')
```

```python
loaded_iforest = load_model('isolation_forest_anomaly_detection')
loaded_iforest
```

```python
save_model(svm_model, 'svm_anomaly_detection')
```

```python
loaded_svm_model = load_model('svm_anomaly_detection')
loaded_svm_model
```
