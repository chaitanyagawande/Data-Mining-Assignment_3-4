```python
import numpy as np
import pandas as pd
from pycaret.datasets import get_data
from pycaret.time_series import TSForecastingExperiment
```

```python
data = get_data('lynx')
data.head()
```

```python
experiment = TSForecastingExperiment()
experiment.setup(data, fh = 12, fold = 3, target = 'Number of Lynx trappings')
```

```python
experiment.plot_model(plot = 'ts')
```

```python
experiment.plot_model(plot = 'diff')
```

```python
experiment.models()
```

```python
create_model = experiment.create_model('naive')
```

```python
predicted_model = experiment.predict_model(create_model)
```

```python
tuned_model = experiment.tune_model(create_model)
```

```python
experiment.plot_model(estimator=tuned_model)
```

```python
finalized_model = experiment.finalize_model(tuned_model)
finalized_model
```

```python
saved_model = experiment.save_model(finalized_model, 'time_series_model_with_exogenuous_variables')
saved_model
```

```python
loaded_model = experiment.load_model('time_series_model_with_exogenuous_variables')
loaded_model
```
