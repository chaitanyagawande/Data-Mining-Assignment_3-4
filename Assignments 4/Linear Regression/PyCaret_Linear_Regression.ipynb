```python
import pandas as pd
data = pd.read_csv("airbnb.csv")
data.head()
```

```python
comp_models = compare_models()
```

```python
plot_model(comp_models, plot = 'residuals')
```

```python
plot_model(comp_models, plot = 'error')
```

```python
plot_model(comp_models, plot = 'feature')
```

```python
evaluate_model(comp_models)
```

```python
pred = predict_model(comp_models)
```
